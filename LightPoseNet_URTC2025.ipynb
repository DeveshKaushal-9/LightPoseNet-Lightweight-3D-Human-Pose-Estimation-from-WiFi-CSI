```python
# ─── Install Required Packages ───────────────────────────────────────────
!pip install -q torch numpy tqdm matplotlib

# ─── Imports ─────────────────────────────────────────────────────────────
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
from tqdm import tqdm
from google.colab import drive

# ─── Mount Google Drive ──────────────────────────────────────────────────
drive.mount('/content/drive')

# ─── Dataset Path ────────────────────────────────────────────────────────
DATA_DIR = '/content/drive/MyDrive/Wi-Pose/processed_data'

# ─── Inspect Pose Data ───────────────────────────────────────────────────
def inspect_pose_data(data_dir, split="train", num_samples=2):
    pose = np.load(os.path.join(data_dir, split, 'pose.npy'), mmap_mode='r')
    for i in range(min(num_samples, len(pose))):
        sample = pose[i]
        print(f"\nSample {i} shape: {sample.shape}")
        print(f"Sample {i} first 3x3 of dim 0:\n{sample[0, :3, :3]}")
        mean_pose = np.mean(sample, axis=(1, 2))  # Alternative transformation
        print(f"Mean pose per dim: {mean_pose}")

inspect_pose_data(DATA_DIR, "train")
inspect_pose_data(DATA_DIR, "test")

# ─── Dataset Class ───────────────────────────────────────────────────────
class WiPoseV73Dataset(Dataset):
    def __init__(self, data_dir):
        self.csi = np.load(os.path.join(data_dir, 'csi.npy'), mmap_mode='r')
        self.pose = np.load(os.path.join(data_dir, 'pose.npy'), mmap_mode='r')
        self.length = len(self.csi)
        # Precompute mean and std for reference (subset for memory)
        self.pose_mean = np.mean(self.pose[:10000])
        self.pose_std = np.std(self.pose[:10000])

    def __len__(self):
        return self.length

    def __getitem__(self, idx):
        # Load and preprocess CSI
        csi = np.array(self.csi[idx]).reshape(9, 30, 5)  # [3, 3, 30, 5] → [9, 30, 5]
        csi = np.abs(csi)
        csi = (csi - np.mean(csi)) / (np.std(csi) + 1e-6)

        # Load and preprocess pose
        pose = np.array(self.pose[idx])  # [3, 18, 18]
        # Alternative transformation: Average per dimension
        pose = np.mean(pose, axis=(1, 2))  # [3]
        pose = np.tile(pose, 18)  # [54] to match 18 joints × 3 coordinates
        pose = (pose - np.mean(pose)) / (np.std(pose) + 1e-6)

        return torch.tensor(csi, dtype=torch.float32), torch.tensor(pose, dtype=torch.float32)

    def denormalize_pose(self, pose):
        # Adjust to physical scale (e.g., 0.5m as a typical body size normalization)
        return pose * 0.5  # Refine with Wi-Pose’s exact scale if available

# ─── SE Block ─────────────────────────────────────────────────────────────
class SEBlock(nn.Module):
    def __init__(self, channels, reduction=8):
        super().__init__()
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channels, channels // reduction),
            nn.ReLU(inplace=True),
            nn.Linear(channels // reduction, channels),
            nn.Sigmoid()
        )

    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.pool(x).view(b, c)
        y = self.fc(y).view(b, c, 1, 1)
        return x * y

# ─── LightPoseNet with SE ────────────────────────────────────────────────
class LightPoseNetSE(nn.Module):
    def __init__(self, output_dim=54):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(9, 32, 3, padding=1), nn.BatchNorm2d(32), nn.ReLU(), SEBlock(32),
            nn.Conv2d(32, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(), SEBlock(64),
            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(), SEBlock(128),
            nn.AdaptiveAvgPool2d((1, 1))
        )
        self.regressor = nn.Sequential(nn.Flatten(), nn.Linear(128, output_dim))

    def forward(self, x):
        return self.regressor(self.encoder(x))

# ─── LightPoseNet Baseline ───────────────────────────────────────────────
class LightPoseNetBaseline(nn.Module):
    def __init__(self, output_dim=54):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(9, 32, 3, padding=1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1))
        )
        self.regressor = nn.Sequential(nn.Flatten(), nn.Linear(64, output_dim))

    def forward(self, x):
        return self.regressor(self.encoder(x))

# ─── Evaluation Metrics ──────────────────────────────────────────────────
def mpjpe(pred, gt):
    pred = pred.view(-1, 18, 3)
    gt = gt.view(-1, 18, 3)
    return torch.mean(torch.norm(pred - gt, dim=2))

def pck(pred, gt, thresh=0.05):
    pred = pred.view(-1, 18, 3)
    gt = gt.view(-1, 18, 3)
    dist = torch.norm(pred - gt, dim=2)
    return (dist < thresh).float().mean()

# ─── Pose Visualization ──────────────────────────────────────────────────
def plot_pose(pose, title="Pose"):
    pose = pose.reshape(18, 3)
    plt.scatter(pose[:, 0], pose[:, 1])
    for i, (x, y) in enumerate(pose[:, :2]):
        plt.text(x, y, str(i), fontsize=8)
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()

# ─── Training Loop ───────────────────────────────────────────────────────
def train_and_evaluate(model, train_dir, test_dir, model_name="model", epochs=20, batch_size=8, lr=1e-3):
    train_dataset = WiPoseV73Dataset(train_dir)
    test_dataset = WiPoseV73Dataset(test_dir)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
    val_loader = DataLoader(test_dataset, batch_size=batch_size, num_workers=2, pin_memory=True)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print("Using device:", device)

    model = model.to(device)
    opt = optim.Adam(model.parameters(), lr=lr)
    crit = nn.MSELoss()

    history = {"loss": [], "val_loss": [], "MPJPE": [], "PCK_05": [], "PCK_10": [], "PCK_15": []}

    for ep in range(epochs):
        model.train()
        running_loss = 0
        for csr, gt in tqdm(train_loader, desc=f"Epoch {ep+1}/{epochs}"):
            csr, gt = csr.to(device), gt.to(device)
            pred = model(csr)
            loss = crit(pred, gt)
            opt.zero_grad()
            loss.backward()
            opt.step()
            running_loss += loss.item()
        avg_train_loss = running_loss / len(train_loader)

        model.eval()
        val_loss = 0
        total_mpjpe = 0
        pck05 = 0
        pck10 = 0
        pck15 = 0
        with torch.no_grad():
            for csr, gt in val_loader:
                csr, gt = csr.to(device), gt.to(device)
                pred = model(csr)
                # Denormalize for metrics
                pred_denorm = train_dataset.denormalize_pose(pred.cpu())
                gt_denorm = train_dataset.denormalize_pose(gt.cpu())
                val_loss += crit(pred, gt).item()
                total_mpjpe += mpjpe(pred_denorm, gt_denorm).item()
                pck05 += pck(pred_denorm, gt_denorm, 0.05).item()
                pck10 += pck(pred_denorm, gt_denorm, 0.1).item()
                pck15 += pck(pred_denorm, gt_denorm, 0.15).item()

        n_batches = len(val_loader)
        history["loss"].append(avg_train_loss)
        history["val_loss"].append(val_loss / n_batches)
        history["MPJPE"].append(total_mpjpe / n_batches)
        history["PCK_05"].append(pck05 / n_batches)
        history["PCK_10"].append(pck10 / n_batches)
        history["PCK_15"].append(pck15 / n_batches)

        print(f"Ep{ep+1} L:{avg_train_loss:.2f}, ValL:{val_loss/n_batches:.2f}, "
              f"MPJPE:{total_mpjpe/n_batches:.2f}, PCK@0.05:{pck05/n_batches:.4f}, "
              f"PCK@0.10:{pck10/n_batches:.4f}, PCK@0.15:{pck15/n_batches:.4f}")

    # Plot training results
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history["loss"], label="Train Loss")
    plt.plot(history["val_loss"], label="Validation Loss")
    plt.title("Loss Curve")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(history["MPJPE"], label="MPJPE")
    plt.plot(history["PCK_05"], label="PCK@0.05")
    plt.plot(history["PCK_10"], label="PCK@0.10")
    plt.plot(history["PCK_15"], label="PCK@0.15")
    plt.title("Validation Metrics")
    plt.xlabel("Epoch")
    plt.ylabel("Metric Value")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Visualize predictions
    sample = next(iter(val_loader))
    csr_sample, gt_sample = sample[0][0].unsqueeze(0).to(device), sample[1][0]
    pred_sample = model(csr_sample)[0].cpu().detach()
    plot_pose(train_dataset.denormalize_pose(gt_sample).numpy(), "Ground Truth")
    plot_pose(train_dataset.denormalize_pose(pred_sample).numpy(), "Prediction")

    torch.save(model.state_dict(), f"{model_name}.pth")
    print(f"Model saved as {model_name}.pth")

# ─── Run Experiments ──────────────────────────────────────────────────────
TRAIN_DIR = os.path.join(DATA_DIR, "train")
TEST_DIR = os.path.join(DATA_DIR, "test")

print("\n--- Running SE Model ---\n")
train_and_evaluate(LightPoseNetSE(), TRAIN_DIR, TEST_DIR, model_name="lightposenet_se", epochs=20)

print("\n--- Running Baseline Model ---\n")
train_and_evaluate(LightPoseNetBaseline(), TRAIN_DIR, TEST_DIR, model_name="lightposenet_baseline", epochs=20)
```
